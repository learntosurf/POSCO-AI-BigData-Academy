### 데이터 Handling_2

#### 데이터 선택

```python
# 랜덤 시드 설정
np.random.seed(1234)

# randn(): 10행 4열의 2차원 배열 생성 
# 배열의 각 요소는 정규 분포에서 무작위로 추출된 값 
df = pd.DataFrame(randn(10, 4), columns=['A', 'B', 'C', 'D'])
df
```

```python
# slice - B열 선택 
df['B']
```

```python
# slice - B, D열 선택 
df[['B', 'D']]
```

```python
# slice - 4~8행 선택 
df[4:9]
```

```python
# loc - 2~8행, A,C열 선택
df.loc[2:8, ['A', 'C']]
```

```python
# loc - A,C열 선택 
df.loc[:, ['A', 'C']]
```

```python
# loc - 3~8행 선택 
df.loc[3:8]
```

```python
# loc - 3행 선택(단일행)
df.loc[3:3]
```

```python
# iloc - 0~6행, A열 선택
df.iloc[0:7, [0]]
```

```python
# iloc - A열 선택 
df.iloc[:, [0]]
```

```python
# iloc - 2~6행 선택
df.iloc[2:7]
```

```python
# iloc - 5행 선택(단일행)
df.iloc[5:6]
```

```python
# query - D>0.5인 행 선택
df.query('D > 0.5')
```

```python
# query - B>D인 행 선택
df.query('B > D')
```

```python
# query - B>D인 행 선택(slice 표현으로)
df[df.B > df.D]
```

```python
# query - A>0.5이고 D<0.0인 데이터 선택 
df.query('A > 0.5 & D < 0.0')
```

```python
# query - A>0.5 혹은 D<0.0인 데이터 선택 
df.query('A > 0.5 | D < 0.0')
```

#### 데이터 결합 (1)

```python
# 데이터 생성
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])

df2 = pd.DataFrame({'E': ['E4', 'E5', 'E6', 'E7'],
                    'F': ['F4', 'F5', 'F6', 'F7'],
                    'G': ['G4', 'G5', 'G6', 'G7'],
                    'H': ['H4', 'H5', 'H6', 'H7']},
                    index=[0, 1, 2, 3])

df3 = pd.DataFrame({'I': ['I8', 'I9', 'I10', 'I11'],
                    'J': ['J8', 'J9', 'J10', 'J11'],
                    'K': ['K8', 'K9', 'K10', 'K11'],
                    'L': ['L8', 'L9', 'L10', 'L11']},
                    index=[0, 1, 2, 3])
```

```python
# concat를 이용한 df1, df2, df3 결합
result = pd.concat([df1, df2, df3], axis=1)
result
```

```python
# axis=0: 행방향 결합
result = pd.concat([df1, df2, df3], axis=0)
result
```
